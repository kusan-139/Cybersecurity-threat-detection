# 02. Model Training
Training the Random Forest and Isolation Forest models.

```
import pandas as pd
import sys
import os

sys.path.insert(0, os.path.abspath('..'))

from src.preprocessor import preprocess
from src.feature_engineering import get_feature_importances, select_top_features
from src.model_trainer import train_random_forest, train_isolation_forest
from src.evaluator import evaluate_model, plot_confusion_matrix
```

## Preprocessing & Feature Engineering

```
df = pd.read_csv('../data/raw/synthetic_network_traffic.csv')

# Preprocess (Scale, encode, split)
data = preprocess(df)
X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']
feature_names = data['feature_names']
```

```
importance_df = get_feature_importances(X_train, y_train, feature_names)
X_train_sel, X_test_sel, selected_features = select_top_features(X_train, X_test, feature_names, importance_df, top_n=20)
```

## Train Models

```
rf_model = train_random_forest(X_train_sel, y_train, n_estimators=50)
print('Random Forest Trained')

if_model = train_isolation_forest(X_train_sel, contamination=0.20)
print('Isolation Forest Trained')
```

## Evaluate

```
rf_metrics = evaluate_model(rf_model, X_test_sel, y_test, 'Random Forest', save_dir=None)

y_pred_rf = rf_model.predict(X_test_sel)
plot_confusion_matrix(y_test, y_pred_rf, 'Random Forest')
```
